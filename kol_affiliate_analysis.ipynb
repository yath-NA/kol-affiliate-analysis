````python

df = pd.read_csv("kol_affiliate_1000.csv")
````

````python
print("Preview Data:")
display(df.head())
````

````python
print("\nInfo Dataset:")
print(df.info())
````

````python
print("\nStatistik Deskriptif:")
display(df.describe(include="all"))
````

````python
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
````

````python
df["id_kol"] = df["id_kol"].astype(str)   # ID tetap string
df["jumlah_followers"] = pd.to_numeric(df["jumlah_followers"], errors="coerce")
df["ctr_%"] = pd.to_numeric(df["ctr_%"], errors="coerce")
df["roi_%"] = pd.to_numeric(df["roi_%"], errors="coerce")
df["engagement_rate_%"] = pd.to_numeric(df["engagement_rate_%"], errors="coerce")
````

````python
df["sosial_media"] = df["sosial_media"].str.strip().str.capitalize()
````

````python
df["partner"] = df["partner"].str.strip().str.title()
````

````python
print("Missing values per kolom:")
print(df.isnull().sum())
````

````python
numeric_cols = ["jumlah_followers", "ctr_%", "roi_%", "engagement_rate_%"]
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].median())
````

````python
print("\nSetelah cleaning:")
display(df.head())
print(df.info())
````

````python
import seaborn as sns
import matplotlib.pyplot as plt
````

````python
kol_per_platform = df["sosial_media"].value_counts()
print("Jumlah KOL per platform:")
print(kol_per_platform)
````

````python
partner_dist = df["partner"].value_counts()
print("\nDistribusi Partner:")
print(partner_dist)
````

````python
print("\nStatistik Followers:")
print(df["jumlah_followers"].describe())
````

````python
ctr_roi_platform = df.groupby("sosial_media")[["ctr_%","roi_%"]].mean().round(2)
print("\nRata-rata CTR & ROI per platform:")
print(ctr_roi_platform)
````

````python
plt.figure(figsize=(6,4))
sns.countplot(x="sosial_media", data=df, palette="Set2")
plt.title("Jumlah KOL per Platform")
plt.xlabel("Platform")
plt.ylabel("Jumlah KOL")
plt.show()
````

````python
plt.figure(figsize=(5,5))
plt.pie(partner_dist, labels=partner_dist.index, autopct="%1.1f%%", colors=["#66b3ff","#ff9999"])
plt.title("Distribusi Partner (Influencer vs Partner Exclusif)")
plt.show()
````

````python
plt.figure(figsize=(8,5))
sns.histplot(df["jumlah_followers"], bins=30, kde=True, color="skyblue")
plt.title("Distribusi Jumlah Followers")
plt.xlabel("Jumlah Followers")
plt.ylabel("Frekuensi")
plt.show()
````

````python
plt.figure(figsize=(8,6))
sns.scatterplot(x="jumlah_followers", y="engagement_rate_%", hue="sosial_media", data=df, alpha=0.6)
plt.title("Followers vs Engagement Rate per Platform")
plt.xlabel("Jumlah Followers")
plt.ylabel("Engagement Rate (%)")
plt.legend(title="Platform")
plt.show()
````

````python
plt.figure(figsize=(8,5))
ctr_roi = df.groupby("sosial_media")[["ctr_%","roi_%"]].mean().reset_index()
ctr_roi.plot(x="sosial_media", kind="bar", figsize=(8,5), color=["#66b3ff","#ff9999"])
plt.title("Rata-rata CTR & ROI per Platform")
plt.xlabel("Platform")
plt.ylabel("Persentase (%)")
plt.xticks(rotation=0)
plt.legend(["CTR (%)", "ROI (%)"])
plt.show()
````

````python
def categorize_tier(followers):
    if followers < 1000:
        return "Nano (<1K)"
    elif followers < 10000:
        return "Micro (1K-10K)"
    elif followers < 100000:
        return "Small (10K-100K)"
    elif followers < 1000000:
        return "Medium (100K-1M)"
    else:
        return "Large (1M+)"

df["followers_tier"] = df["jumlah_followers"].apply(categorize_tier)
````

````python
plt.figure(figsize=(7,5))
sns.countplot(x="followers_tier", data=df, order=["Nano (<1K)","Micro (1K-10K)","Small (10K-100K)","Medium (100K-1M)","Large (1M+)"],
              palette=["#90ee90","#66b3ff","#ffcc66","#cc99ff","#ff6666"])
plt.title("Distribusi KOL berdasarkan Followers Tier")
plt.xlabel("Tier")
plt.ylabel("Jumlah KOL")
plt.show()
````

````python
plt.figure(figsize=(8,6))
sns.heatmap(df[["jumlah_followers","ctr_%","roi_%","engagement_rate_%"]].corr(), annot=True, cmap="Blues", fmt=".2f")
plt.title("Korelasi Antar Variabel Numerik")
plt.show()
````

````python
plt.figure(figsize=(8,6))
sns.scatterplot(x="jumlah_followers", y="roi_%", hue="sosial_media", data=df, alpha=0.6)
plt.title("Followers vs ROI per Platform")
plt.xlabel("Jumlah Followers")
plt.ylabel("ROI (%)")
plt.legend(title="Platform")
plt.show()
````
